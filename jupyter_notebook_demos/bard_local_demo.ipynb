```python
# !pip install bardapi
import os
import time
from bardapi import Bard
```

```python
# key = os.environ['_BARD_API_KEY']
key = '[Copy your Bard API Key Here]'
bard = Bard(token=key)
```

```python
def get_bard_response(bard, image_path, input_text):
    try:
        image = open(image_path, 'rb').read() # (jpeg, png, webp) are supported.
        response = bard.ask_about_image(input_text, image)
        response = response['content'].strip()
        return response
    except Exception as e:
        print(e)
    return ""
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

def display_image(image_path):
    if os.path.exists(image_path): 
        img = imread(image_path)

        fig = plt.figure()
        fig.set_dpi(50)
        
        plt.imshow(img)
        plt.axis('off')
        plt.show()
```

## Demo 1

```python
image_path = "../data/images/852.jpg"

query = "Hint: Please answer the question requiring an integer answer and provide the final value, e.g., 1, 2, 3, at the end.\nQuestion: Which number is missing?"

response = get_bard_response(bard, image_path, query)

display_image(image_path)
print(query)
```

```python
print(response)
```
